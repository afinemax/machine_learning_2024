My neighbor has two pets. Assume that the choice to own a cat or a dog is like a coin flip: the two options are mutually exclusive and exhaustive, per pet.

a. Which possibility is the most likely, a priori, given the assumptions above: two cats, two dogs, or one cat and one dog?

b. Suppose I ask my neighbor if he has any dogs, and he says yes. What is the probability that one pet is a cat?

c. Suppose that instead I see one of his pets run by in his house and its a dog. What is the probability that the other pet is a cat?

Use probability tables to justify your results.


### Solution 


| Outcome        | Probability |
|----------------|-------------|
| One cat, One dog | $\frac{2}{4}$ |
| Two dogs       | $\frac{1}{4}$ |
| Two cats       | $\frac{1}{4}$ |

### A.
A priori, given these assumptions, the most likely outcome is one cat and one dog. This option occurs 
$\frac{2}{4}$ times while two dogs happen  $\frac{1}{4}$ of the time, and two cats occur $\frac{1}{4}$ of the time.

### B.
Assuming the Neighbor has two pets, the probability my neighbor has a cat, given he has a dog is...

This is a Bayes' theorem question! Recall: 
$$p(a|b) = \frac{p(b|a) p(a)}{p(b)} = \frac{p(a \cup b)}{p(b)}$$

In our case, let $b$ be dogs and $a$ be cats:

- $p(b) = 0.75$

- $p(a) = 0.75$

- $p(a \cup b) = 0.5$

So 
$$p(b|a) = \frac{p(a \cup b)}{p(b)} = \frac{2}{3}$$

Answer $= \frac{2}{3}$

### C.
This is the same problem as B, (and is the same answer if the pets are inverted), so 
$\frac{2}{3}$.

### 1.2 Monty Hall Problem (2 points)

*Topics: joined and marginal distributions, probability tables*

On a game show, a contestant is told the following rules:


> There are three doors, labelled 1, 2, 3. A single prize has been hidden behind one of them. You get to select one door. Initially your chosen door will not be opened. Instead, the gameshow host will open one of the other two doors, and *he will do so in such a way as not to reveal the prize.* For example, if you first choose door 1, he will then open one of doors 2 and 3, and it is guaranteed that he will choose which one to open so that the prize will not be revealed. At this point, you will be given a fresh choice of door: you can either stick with your first choice, or you can switch to the other closed door. All the doors will then be opened and you will receive whatever is behind your final choice of door.

Imagine that the contestant chooses door 1 first; then the gameshow host opens door 3, revealing nothing
behind the door, as promised. Should the contestant (a) stick with door 1, or (b) switch to door 2, or (c) does it make no difference? You may assume that initially, the prize is equally likely to be behind any of the 3 doors.

Use probability tables to justify your answer.

### Solution 


| Behind door 1 | Behind door 2 | Behind door 3 | Result if staying at door #1 | Result if switching to the door offered |
|---------------|---------------|---------------|------------------------------|----------------------------------------|
| No Prize        | No Prize          | Prize           | Wins Nothing                   | Wins Prize                              |
| No Prize        | Prize         | No Prize         | Wins Nothing                  | Wins Prize                             |
| Car           | No Prize        | No Prize          | Wins Prize                     | Wins Nothing                             |


The solution to the Monty Hall problem is counterintuitive but can be explained using probability:

1. When you initially choose one door, you have a 1/3 chance of picking the prize and a 2/3 chance of picking a no prize.
2. When the host opens one of the other doors to reveal a non-prize, the probabilities do not change!. The door you originally picked still has a 1/3 chance of hiding the prize, but now the remaining unopened door has a 2/3 chance of hiding the prize. This is because probabilitiy is conserved, the remaining door now holds 2 inital doors worth of probabilty. 
3. Therefore, it is always advantageous to switch doors. You double your chances of winning the car if you switch. Switching gives you a 2/3 chance of winning!

### 1.4 Sampling from a Probabilistic Model (2 points)


*Topic: frequentist probabilities*

Let's say we have a probabilistic model with

| X | p(X) |
|------|------|
| Sun (0)   | 0.2 |

a. Write a sampler that randomly samples from $p(X)$ values X = 0, 1, 2. In your solution, you are allowed to use the function `numpy.random.random()` but other than this, do not use any other imported functions.

```python
from numpy.random import random # Return random floats in the half-open interval [0.0, 1.0).
import numpy as np # glory to numpy
import matplotlib.pyplot as plt


# Probabilities
probabilities = [0.2, 0.3, 0.5]

def return_random_sample(probabilities):
    '''
    A sampler that randomly samples from a discrete probability distribution.

    Parameters:
        probabilities (list): List of probabilities corresponding to each possible outcome.

    Returns:
        int or None: Sampled value based on the probabilities provided. Returns None if probabilities are not provided.

    Example:
        >>> return_random_sample([0.1, 0.2, 0.7])
        2
    '''

    # Generate a random number between 0 and 1
    random_number = np.random.random()

    # Initialize sampled value
    sampled_value = None

    # Iterate over the probability table 
    # Chooses the assocaited X for prob(x)
    # Once the cumulative probability exceeds the random number, it selects the corresponding value of X.
    cumulative_prob = 0.0
    for i in range(len(probabilities)):
    
        cumulative_prob += probabilities[i]
        if random_number < cumulative_prob: # Check if prob is chosen
            sampled_value = i # Get X for prob(x)
            break
    
    return sampled_value # one value


# Test
############################################################################################
# Number of samples
num_samples = 10000

# Counters for sampled values
sample_counts = [0, 0, 0]

# Perform sampling
for _ in range(num_samples):
    sampled_value = return_random_sample(probabilities)
    sample_counts[sampled_value] += 1

# Calculate frequencies
frequencies = [count / num_samples for count in sample_counts]

# Expected probabilities
expected_probabilities = probabilities

# Print results
print("Sampled Frequencies:", frequencies)
print("Expected Probabilities:", expected_probabilities)
```

a. Implement a sampling algorithm that samples $(X, Y)$ pairs.

```python
probability_table = [
        (("Hot", "Sun"), 0.4),
        (("Hot", "Rain"), 0.1),
        (("Cold", "Sun"), 0.2),
        (("Cold", "Rain"), 0.3)
    ]

def return_random_sample_joint(probability_table):
    """
    Return a random sample from a joint probability distribution defined by the given probability table.

    Parameters:
    - probability_table (list of tuples): A list of tuples where each tuple represents a joint outcome and its probability.
                                          Each tuple should have the format ((X, Y), p), where X and Y are the outcomes of
                                          random variables X and Y respectively, and p is the probability of the joint outcome.

    Returns:
    - tuple or None: A randomly sampled joint outcome from the probability distribution defined by the probability table.
                     Returns None if the probability table is empty or if the random sample falls outside the defined
                     probability distribution.

    Example:
    >>> probability_table = [(("Hot", "Sun"), 0.4), (("Hot", "Rain"), 0.1), (("Cold", "Sun"), 0.2), (("Cold", "Rain"), 0.3)]
    >>> return_random_sample_joint(probability_table)
    ('Cold', 'Rain')
    """
    
    # Calculate cumulative probabilities
    cumulative_probs = [0]
    for _, prob in probability_table:
        cumulative_probs.append(cumulative_probs[-1] + prob)
    
    # Generate a random number
    rand_num = np.random.random()
    
    # Find the corresponding pair based on the random number
    for i in range(len(cumulative_probs) - 1):
        if cumulative_probs[i] <= rand_num < cumulative_probs[i + 1]:
            return probability_table[i][0]
    
    # If no pair is found, return None
    return None

# Test the function
####################################################################################
# Sample (X, Y) pairs
num_samples = 1000
samples = [return_random_sample_joint(probability_table) for _ in range(num_samples)]

# Count the occurrences of each (X, Y) pair
count = {}
for sample in samples:
    if sample in count:
        count[sample] += 1
    else:
        count[sample] = 1


print('Note: Order of the tables are not the same\n')

# Print observed frequencies
print("Observed frequencies:")
for key, value in count.items():
    print(key, ":", value / num_samples)

# Compare with expected probabilities from the table
print("\nExpected probabilities from the table:")
for entry in probability_table:
    print(entry[0], ":", entry[1])

```

```python
import numpy as np
import matplotlib.pyplot as plt

def mvgauss(x, mu, sigma):
    """
    Compute the multivariate Gaussian PDF at point x with mean mu and covariance sigma.
    
    Parameters:
        x (numpy array): Point at which to evaluate the PDF.
        mu (numpy array): Mean vector of the multivariate Gaussian.
        sigma (numpy array): Covariance matrix of the multivariate Gaussian.
        
    Returns:
        float: Value of the multivariate Gaussian PDF at point x.
    """
    n = len(mu) # number of dimensions 
    det_sigma = np.linalg.det(sigma)
    inv_sigma = np.linalg.inv(sigma)
    exponent = -0.5 * np.dot(np.dot((x - mu).T, inv_sigma), (x - mu))
    coefficient = 1 / (np.sqrt((2 * np.pi) ** n * det_sigma))
    return coefficient * np.exp(exponent)

# Define parameters for plotting
x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
X, Y = np.meshgrid(x, y) # 2d plotting grid

# Define means and covariance matrices
mu1 = np.array([5, 5])
sigma1 = np.array([[1, 0], [0, 1]])

mu2 = np.array([5, 5])
sigma2 = np.array([[1, -0.5], [-0.5, 1]])

mu3 = np.array([5, 5])
sigma3 = np.array([[1, 0.7], [0.7, 1]])

# Evaluate PDFs
Z1 = np.array([mvgauss(np.array([xi, yi]), mu1, sigma1) for xi, yi in zip(np.ravel(X), np.ravel(Y))])
Z1 = Z1.reshape(X.shape)

Z2 = np.array([mvgauss(np.array([xi, yi]), mu2, sigma2) for xi, yi in zip(np.ravel(X), np.ravel(Y))])
Z2 = Z2.reshape(X.shape)

Z3 = np.array([mvgauss(np.array([xi, yi]), mu3, sigma3) for xi, yi in zip(np.ravel(X), np.ravel(Y))])
Z3 = Z3.reshape(X.shape)

# Plot 2D contour plots
plt.figure(figsize=(15, 5))

plt.subplot(131)
plt.contour(X, Y, Z1)
plt.title('Contour Plot - $\mu_1$, $\Sigma_1$', fontsize=15)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)

plt.subplot(132)
plt.contour(X, Y, Z2)
plt.title('Contour Plot - $\mu_2$, $\Sigma_2$', fontsize=15)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)

plt.subplot(133)
plt.contour(X, Y, Z3)
plt.title('Contour Plot - $\mu_3$, $\Sigma_3$', fontsize=15)
plt.xlabel('X', fontsize=15)
plt.ylabel('Y', fontsize=15)

plt.show()

# Plot 3D surface plots
fig = plt.figure(figsize=(15, 5))

ax1 = fig.add_subplot(131, projection='3d')
ax1.plot_surface(X, Y, Z1, cmap='viridis')
ax1.set_title('Surface Plot - $\mu_1$, $\Sigma_1$', fontsize=15)
ax1.set_xlabel('X', fontsize=15)
ax1.set_ylabel('Y', fontsize=15)
ax1.set_zlabel('Z', fontsize=15)

ax2 = fig.add_subplot(132, projection='3d')
ax2.plot_surface(X, Y, Z2, cmap='viridis')
ax2.set_title('Surface Plot - $\mu_2$, $\Sigma_2$', fontsize=15)
ax2.set_xlabel('X', fontsize=15)
ax2.set_ylabel('Y', fontsize=15)
ax2.set_zlabel('Z', fontsize=15)

ax3 = fig.add_subplot(133, projection='3d')
ax3.plot_surface(X, Y, Z3, cmap='viridis')
ax3.set_title('Surface Plot - $\mu_3$, $\Sigma_3$', fontsize=15)
ax3.set_xlabel('X', fontsize=15)
ax3.set_ylabel('Y', fontsize=15)
ax3.set_zlabel('Z', fontsize=15)

plt.show()
```
